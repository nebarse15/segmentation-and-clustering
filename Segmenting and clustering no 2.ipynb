```python
import pandas as pd #library for data analysis
import numpy as np #library to handle data in vectoral manner
import random #library for number generation
import requests #library to handle request
from bs4 import BeautifulSoup
from IPython.core.display import HTML
from IPython.display import display_html
```

```python
#scraping the wikipedia page for the postal codes
List_url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
Source = requests.get(List_url).text
```

```python
city_soup = BeautifulSoup(Source,'lxml')
```

```python
Table = str(city_soup.table)
```

```python
display_html(Table,raw = True)
```

```python
df = pd.read_html(Table)
df1 = df[0]
df1.head(10)
```

```python
#Drop the rows in which borough is not assigned
df2 = df1[df1.Borough != 'Not assigned']
```

```python
# The neighborhoods with the same postal codes will be grouped
dfg = df2.groupby(['Postal Code', 'Borough'],sort = False).agg(','.join)
dfg.reset_index(inplace = True)
```

```python
#Neighborhoods which are not assigned will be replaced by the name of the boroughs
dfg['Neighborhood'] = np.where(dfg['Neighborhood'] == 'Not assigned', dfg['Borough'],dfg['Neighborhood'])
dfg
```
